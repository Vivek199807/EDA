##4. Data Vizualization, Storytelling & Experimenting with charts

1)The distribution of Airbnb listings across different neighbourhood groups
in New York City

```python
#making a column of total number of listing in different neighbourhood groups
total_counts=airbnb_eda_df['neighbourhood_group'].value_counts().reset_index()
total_counts.columns=['neighbourhood_group','count']
total_counts
```

```python
# Set the theme for the plot
sns.set_theme(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x=airbnb_eda_df['neighbourhood_group'] , palette='viridis')

# Add titles and labels
plt.title('Neighbourhood groups distribution', fontsize=20)
plt.xlabel('Neighbourhood groups', fontsize=14)
plt.ylabel('Listing counts', fontsize=14)

# Customize the appearance
sns.despine(left=True)  # Remove the left spine
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display the plot
plt.show()
```

**Insights**



1)Brooklyn & Manhattan have the highest number of listings, with over 20,000 listings respectively.

2)Queens & the Bronx have significantly lower listings Brooklyn as well as Manhattan, with slightly more than 5000 and 1000 listings respectively.

3)Staten Island has the lowest number of listings, with only around 350 - 400 listings.

4)The distribution of listings across the different neighborhood groups is skewed, with a concentration of listings in Manhattan and Brooklyn.

5)Despite being larger in size, the neighborhoods in Queens, the Bronx, and Staten Island have fewer listings on Airbnb compared to Manhattan, which has a smaller geographical area.

6)This could suggest that the demand for Airbnb rentals is higher in Manhattan compared to the other neighborhoods, leading to a higher concentration of listings in this area thus to cater those demands higher number of property owners in this area are listing their properties on Airbnb.
  



(2)Variation of average prices of Airbnb rentals by neighbourhood groups

```python
group_avg=airbnb_eda_df.groupby('neighbourhood_group')['price'].mean().reset_index()
group_avg
```

```python
# Create the line plot
plt.figure(figsize=(8, 6))
sns.lineplot(x='neighbourhood_group', y='price', data=airbnb_eda_df)

# Add titles and labels
plt.title('Avg Price Variations', fontsize=20)
plt.xlabel('Neighbourhood group', fontsize=14)
plt.ylabel('Price', fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Display the plot
plt.show()
```

**Insights**

1)The average price of a listing in New York City varies significantly across different neighborhoods, with Manhattan having the highest ~ 197 dollars/day average price and the Bronx having the lowest near ~ 88 dollars/day.

2)The change in trend depicts that Manhattan is the centre of the city or we can say it's a prime location, with Bronx being suitated at outskirts of the city.

3)The average price in Queens and Staten Island is relatively similar, despite being in different parts of the city.

4)The data suggests that the overall cost of living in New York City is higher in the center of the city (Manhattan) compared to the outer parts of city. This is likely due to the fact that Manhattan is the most densely populated and commercially important area, and therefore has higher demand for housing in the centrally located neighborhoods.

3) What is the distribution of room types (e.g., entire home, private room, shared room) across different neighborhoods?

```python
plt.rcParams['figure.figsize'] = (8, 5)
ax = sns.countplot(y='room_type', hue='neighbourhood_group', data=airbnb_eda_df, palette='bright')
total = len(airbnb_eda_df['room_type'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.title('count of each room types in entire NYC', fontsize='15')
plt.xlabel('Room counts', fontsize='14')
plt.xticks(rotation=90)
plt.ylabel('Rooms', fontsize='14')
plt.show()
```

Insightss -->

1) Manhattan has more listed properties with Entire home/apt around 24.6% of total listed properties followed by Brooklyn with around 19.5%.

2) Private rooms are more in Brooklyn as in 21.9% of the total listed properties followed by Manhattan with 16.9% of them. While 7.3% of private rooms are from Queens.

3) Very few of the total listed have shared rooms listed on Airbnb where there's negligible or almost very rare shared rooms in Staten Island and Bronx.

4)We can infer that Brooklyn,Queens,Bronx has more private room types while Manhattan which has the highest no of listings in entire NYC has more Entire home/apt room types.

(9) Number Of Active Hosts Per Location Using Line Chart

```python
hosts_per_location =airbnb_eda_df.groupby('neighbourhood_group')['host_id'].count().reset_index()
hosts_per_location.columns = ['Neighbourhood_Groups', 'Host_counts']
hosts_per_location
```

```python
hosts_per_location = airbnb_eda_df.groupby('neighbourhood_group')['host_id'].count()
locations = hosts_per_location.index
host_counts = hosts_per_location.values
plt.figure(figsize=(12, 5))
plt.plot(locations, host_counts, marker='o', ms=12, mew=4, mec='r')
plt.title('Number of Active Hosts per Location', fontsize='15')
plt.xlabel('Location', fontsize='14')
plt.ylabel('Number of Active Hosts', fontsize='14')
plt.show()
```

Insightss -->

Manhattan has the largest number of hosts with 19501,Brooklyn has the second largest number of hosts with 19415.

After that Queens with 5567 and the Bronx with 1070. while Staten Island has the fewest with 365.

Brooklyn and Manhattan have the largest number of hosts, with more than double the number of hosts in Queens and more than 18 times the number of hosts in the Bronx.

(10) Correlation Heatmap Visualization

```python
corr = airbnb_eda_df[['id','host_id','latitude','longitude','price','minimum_nights','number_of_reviews','availability_365']].corr()
corr
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(corr, cmap='BrBG',annot=True)
plt.show()
```

Insightss -->

There is a moderate positive correlation (0.58) between the host_id and id columns, which suggests that hosts with more listings are more likely to have unique host IDs.

There is a weak positive correlation (0.17) between the price column and the calculated_host_listings_count column, which suggests that hosts with more listings tend to charge higher prices for their listings.

There is a moderate positive correlation (0.23) between the calculated_host_listings_count column and the availability_365 column, which suggests that hosts with more listings tend to have more days of availability in the next 365 days.

There is a strong positive correlation (0.58) between the number_of_reviews column and the reviews_per_month column, which suggests that listings with more total reviews tend to have more reviews per month.
